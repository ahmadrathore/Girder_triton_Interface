```python
import matplotlib.pyplot as plt
import numpy as np


labels = ['']
read_time_no_cache = [157.92]
read_time_cache = [124.00]



x = np.arange(len(labels))  # the label locations
width = 0.1  # the width of the bars




fig, ax = plt.subplots()

rects1 = ax.bar(x - width/2, read_time_no_cache, width, label='Read_time_No_cache')
rects2 = ax.bar(x + width, read_time_cache, width, label='Read_time_cache_enabled')





# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Time (seconds)')
ax.set_xlabel('Read Time with/Without Cache')
ax.set_title('Time (sec) for files reading with/without Cache memory enabled (.tfr, size=14G)')
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=1, fontsize=10)
ax.bar_label(rects2, padding=1, fontsize=10)
# ax.bar_label(rects3, padding=5, fontsize=8)



fig.tight_layout()

plt.show()
```
